```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import datetime

df = pd.read_csv('wifi_test.csv', index_col=0)
df = df.reset_index()
df['test_time']=df['index'].apply(lambda x: datetime.datetime.strptime(x, '%Y%m%d_%H:%M:%S'))
```

```python
df.shape
```

## set datetime range

```python
dt_start = datetime.datetime.strptime('20211030_7:00:00', "%Y%m%d_%H:%M:%S")
dt_end = datetime.datetime.strptime('20211031_8:00:00', "%Y%m%d_%H:%M:%S")

df = df[(df.test_time>dt_start)&(df.test_time<dt_end)]

df.shape
```

```python
plt.style.use('seaborn-whitegrid')

fig = plt.figure(figsize=(16,6))
ax = fig.add_subplot(111)
for i in df[['down','up']]:
    ax.scatter(x=df.test_time.values, y=df[i], alpha=0.35, label=i)

ax.set_xlim(df.test_time.min() - datetime.timedelta(hours=1),
            df.test_time.max() + datetime.timedelta(hours=1))
ax.set_xlabel('Time')
ax.set_ylabel('WiFi Speed (Mbps)')
ax.legend()

fig = plt.figure(figsize=(16,5))
ax2 = fig.add_subplot(121)
ax3 = fig.add_subplot(122)
for i in df[['down','up']]:
    ax2.hist(df[i].fillna(0), bins=50, edgecolor='black', alpha=0.4,density=True)
    sns.kdeplot(df[i].fillna(0))
ax2.set_xlabel('WiFi Speed (Mbps)')
ax2.legend()
ax3.set_xlabel('WiFi Speed (Mbps)');
```
